**Categorizing 'Vote_Average' column**

We would cut the **Vote_Average** column values and make 4 categories: **popular**, **average** , **below_avg** , **not_popular** to describe it more using **categorize_col()**

```python
def categorize_col(df,col,labels):

    edges = [df[col].describe()["min"],
            df[col].describe()["25%"],
            df[col].describe()["50%"],
            df[col].describe()["75%"],
            df[col].describe()["max"]]
    df[col] = pd.cut(df[col],edges,labels=labels,duplicates = "drop")
    return df
```

```python
labels = ["not_popular","below_avg","average","popular"]

categorize_col(df, "Vote_Average", labels)

df["Vote_Average"].unique()                                    # calling all the unique values after being categorized 
```

```python
# checking for missing values after changes in 'Vote_Average' column 

df.isnull().sum()
```

### We'd split genres into a list and then explode dataframe to have only one genre per row for each movie 

```python
df["Genre"] = df["Genre"].str.split(", ")

df = df.explode("Genre").reset_index(drop=True)
df.head(10)
```

```python
# casting column into category 

df["Genre"] = df["Genre"].astype("category")
df["Genre"].dtypes
```

### What is the most frequent genre of movies released on Netflix ?

```python
# displays some specifications about 'Genre' column 

df["Genre"].describe()
```

```python
sns.catplot(data=df, y="Genre", kind="count",
           order = df["Genre"].value_counts().index,
           color = "#4287f5")
plt.title("Genre Column Distribution")
plt.show()                                             
```
